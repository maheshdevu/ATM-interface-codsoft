```python
def split_word(word):
    return [(word[:i], word[i:]) for i in range(len(word) + 1)]
split_word('jatin')
```

```python
def delete_word(word):
    return [l + r[1:] for l,r in split_word(word) if r]
delete_word('jatin')
```

```python
def swap_word(word):
    return [l + r[1] + r[0] + r[2:] for l, r in split_word(word) if len(r)>1]
swap_word('jatin')
```

```python
import string
def replace_word(word):
    letters = string.ascii_lowercase
    return [l + c + r[1:] for l, r in split_word(word) if r for c in letters]
replace_word('jatin')
```

```python
def insert_word(word):
    letters = string.ascii_lowercase
    return [l + c + r for l, r in split_word(word) for c in letters]
insert_word('jatin')
```

```python
def word_combinations(word):
    return set(delete_word(word) + swap_word(word) + replace_word(word) + insert_word(word))
word_combinations('jatin')
```

```python
def all_word_combinations(word):
    return set(e2 for e1 in word_combinations(word) for e2 in word_combinations(e1))
all_word_combinations('jatin')
```

```python
def correct_spelling(word, vocabulary, word_probabilities):
    if word in vocabulary:
        print(f"{word} is already correctly spelled")
        return

    suggestions = word_combinations(word) or all_word_combinations(word) or [word]
    best_guesses = [w for w in suggestions if w in vocabulary]
    return [(w, word_probabilities[w]) for w in best_guesses]
```

```python
correct_spelling('volcabulary',vocabulary,word_probabilities)
```

```python
def generate_combinations(word):
    letters = string.ascii_lowercase
    splitword = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    deleteword = [l + r[1:] for l,r in splitword if r]
    swapword = [l + r[1] + r[0] + r[2:] for l, r in splitword if len(r)>1]
    replaceword = [l + c + r[1:] for l, r in splitword if r for c in letters]
    insertword = [l + c + r for l, r in splitword for c in letters]

    return set(deleteword + swapword + replaceword + insertword)
```

```python
generate_combinations('how')
```

```python
def generate_all_combinations(word):
    return set(e2 for e1 in generate_combinations(word) for e2 in generate_combinations(e1))
generate_all_combinations('how')
```

```python
def correct_spelling(word, vocabulary, word_probabilities):
    if word in vocabulary:
        print(f"{word} is already correctly spelt")
        return

    suggestions = generate_combinations(word) or generate_all_combinations(word) or [word]
    best_guesses = [w for w in suggestions if w in vocabulary]
    return [(w, word_probabilities[w]) for w in best_guesses]
```
